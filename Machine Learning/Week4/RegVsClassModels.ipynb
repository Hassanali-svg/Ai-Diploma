```python
# Importing necessary libraries
from sklearn import tree
from sklearn import ensemble
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
import numpy as np
```

```python
def RegVsClass(df,label_col_name):
    y=df[label_col_name]
    x=df.drop([label_col_name],axis=1)
    
    X_train,X_test,y_train,y_test=train_test_split(x,y)

    # Reg 
    DT_reg_Model=tree.DecisionTreeRegressor()
    RF_reg_Model=ensemble.RandomForestRegressor()
        
        
    # Class   
    DT_cls_Model=tree.DecisionTreeClassifier()
    RF_cls_Model=ensemble.RandomForestClassifier()


    DT_reg_Model.fit(X_train,y_train)
    RF_reg_Model.fit(X_train,y_train)
    RF_cls_Model.fit(X_train,y_train)
    DT_cls_Model.fit(X_train,y_train)


    print('----REG MODELS----')
    print('DT')
    print(f'R2 Score Acc= {metrics.r2_score(y_test,DT_reg_Model.predict(X_test))}')
    print('RF')
    print(f'R2 Score Acc= {metrics.r2_score(y_test,RF_reg_Model.predict(X_test))}')

    print('========================================================================')
    
    print('----REG MODELS----')
    print('DT')
    print(f'R2 Score Acc= {metrics.accuracy_score(y_test,DT_cls_Model.predict(X_test))}')
    print('RF')
    print(f'R2 Score Acc= {metrics.accuracy_score(y_test,RF_cls_Model.predict(X_test))}')

```

# Social_Network_Ads 

```python
df=pd.read_csv('Social_Network_Ads (1).csv')
df
```

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
```

```python
RegVsClass(df,'Purchased')
```
