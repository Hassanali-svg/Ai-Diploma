```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Position_Salaries (1).csv")
```

```python
df
```

```python
df.info()
```

```python
df.loc[:,'Salary']
```

```python
X=df[['Level']]
y=df[['Salary']]
plt.scatter(X,y)
```

```python
from sklearn.tree import DecisionTreeRegressor


model=DecisionTreeRegressor(random_state=0,max_depth=2)
```

```python
model.fit(X,y)
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(y))
```

```python
from sklearn.tree import plot_tree
```

```python
fig=plt.figure(figsize=[20,20])
v=plot_tree(model,feature_names=df.columns,filled=True)
```

# Cardataet

```python
df=pd.read_csv('Carseats.csv')
```

```python
df.info()
```

```python
print(df['Urban'].unique()
)

print(df['ShelveLoc'].unique())

print(df['US'].unique())

```

```python
df['Urban']=df['Urban'].map({'Yes':0,'No':1})
df['ShelveLoc']=df['ShelveLoc'].map({'Bad':0,'Good':1,'Medium':2})
df['US']=df['US'].map({'Yes':0,'No':1})
```

```python
df.columns
```

```python
X=df[['Sales', 'CompPrice', 'Income', 'Advertising', 'Population',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']]

y=df['Price']

```

```python
model=DecisionTreeRegressor(random_state=0,max_depth=10)
model.fit(X,y)
```

```python
fig=plt.figure(figsize=[30,10],dpi=600)
v=plot_tree(model,feature_names=df.columns,filled=True)
```
